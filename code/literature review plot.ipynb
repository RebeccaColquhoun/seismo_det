```python
import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from cmcrameri import cm 
```

```python
a = [6.3,4.76,7.0,6.1,4.7,7.14,6.83,6.66]
b = [7.1,6.52,5.9,6.7,4.8,5.928,6.36,5.22]
papers = ['allen + kanamori 2003 M3-5',
          'me', 
          'allen + kanamori 2003 M 4.5+',
         'lockman + allen 2007 M<5',
          'lockman + allen 2007 M>=5',
         'olson + allen 2005',
         'tsang et al 2007',
         'wurman et al 2007']
min_mag = [3,3.4,4.5,3.5,5,3,3,3]
max_mag = [5,'?','?',5,6.8,8.3,7.3,7.1]
color = [1,2,1,3,3,4,5,6]
```

```python
df = pd.DataFrame({'papers':papers,'a':a,'b':b,'minmag':min_mag, 'maxmag':max_mag, 'color':color})
```

```python
cmap = cm.hawaii
```

```python
cmap(360)
```

```python
for i in range(0, len(df)):
    row = df.iloc[i]
    if row.maxmag!='?':
        x = np.arange(row.minmag,row.maxmag,0.1)
    else:
        x = np.arange(row.minmag, 8.3,0.1)
    plt.plot(x, (1/row.a)*x-(row.b/row.a), label = row.papers, color = cmap((row.color-1)*60))
plt.legend()
plt.ylabel('log10(tpmax)')
plt.xlabel('magnitude')
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
```
